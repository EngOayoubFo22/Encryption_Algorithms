```python
# Caesar Encryption encoding
print("Welcome to Caesar encoder")
tst = input("Enter string to encrypt: ")
key = int(input("Enter key: "))
key %= 26 # Ensure key is within 0-25

new = ''
for char in tst:
    if 'a' <= char <= 'z':
        start = ord('a')
        shifted_char_code = (ord(char) - start + key) % 26
        new += chr(start + shifted_char_code)
    elif 'A' <= char <= 'Z':
        start = ord('A')
        shifted_char_code = (ord(char) - start + key) % 26
        new += chr(start + shifted_char_code)
    else:
        # Keep non-alphabetic characters unchanged
        new += char
print(new)
```

```python
#Caesar Encryption Decoding
print("Welcome to Caesar decoder")
tst = input("Enter string to decrypt: ")
key = int(input("Enter key: "))
key %= 26 # Ensure key is within 0-25
new = ''

for char in tst:
    if 'a' <= char <= 'z':
        start = ord('a')
        #decryption logic: (0-indexed char position - key + 26) % 26
        shifted_char_code = (ord(char) - start - key + 26) % 26
        new += chr(start + shifted_char_code)
    elif 'A' <= char <= 'Z':
        start = ord('A')
        shifted_char_code = (ord(char) - start - key + 26) % 26
        new += chr(start + shifted_char_code)
    else:
        # Keep non-alphabetic characters unchanged
        new += char
print(new)
```

```python
#Caesar Brute Force Decryption

def caesar_brute_force_decrypt(word):
    """
    Attempts to decrypt a Caesar cipher by trying all 26 possible shifts.
    Prints each potential decryption.
    """
    print(f"\nAttempting to decrypt: '{word}'")
    print("-" * (len(word) + 25))

    # We try shifts from 0 to 25 (26 possibilities)
    # A shift of '1' means 'a' becomes 'b', so to reverse 'b' back to 'a', we need a shift of -1.
    # So, for decryption, our effective shift will be -i.
    for i in range(26): # i represents the *original* encryption shift that we are trying to reverse
        # For each possible shift, we'll build a potential decrypted message
        potential_decryption = []

        # This is the actual shift amount for decryption
        # If the original encryption shift was 'i', then the decryption shift is '-i'
        # To handle modulo correctly for negative numbers, a common trick is (x % n + n) % n
        # or simply work with (26 - i) as the decryption shift if i was the encryption shift.
        decryption_shift = (26 - i) % 26 # This correctly gets the reverse shift

        for char in word:
            if 'a' <= char <= 'z':
                start = ord('a')
                # (ord(char) - start + decryption_shift) gives the new 0-25 position
                shifted_char_code = (ord(char) - start + decryption_shift) % 26
                potential_decryption.append(chr(start + shifted_char_code))
            elif 'A' <= char <= 'Z':
                start = ord('A')
                shifted_char_code = (ord(char) - start + decryption_shift) % 26
                potential_decryption.append(chr(start + shifted_char_code))
            else:
                # Keep non-alphabetic characters unchanged
                potential_decryption.append(char)

        # Print each potential decryption along with the shift it corresponds to
        print(f"Shift -{i:2}: {''.join(potential_decryption)}") # Using f-string for formatting

    print("-" * (len(word) + 25))


# --- How to use it ---
string_to_decrypt = input("Enter string to decrypt: ")
caesar_brute_force_decrypt(string_to_decrypt)

# Example: If you encrypt "hello" with a shift of 3, you get "khoor"
# Enter string to decrypt: khoor
# Look for the readable output in the list.
```
